```python
# Extracting audio features using Librosa 
# Identifying valid train and test candidate ID's
import os
import librosa
import numpy as np
import soundfile as sf
from pydub import AudioSegment

def extract_audio_features(candidate_ids, sounds_folder):
    features = []
    valid_ids = []
    for candidate_id in candidate_ids:
        audio_path = os.path.join(sounds_folder, f"{candidate_id}.wav")
        if not os.path.exists(audio_path):
            print(f"Skipping {candidate_id}: File not found")
            continue

        try:
            with sf.SoundFile(audio_path) as f:
                if f.format != "WAV":
                    print(f"Converting {candidate_id} to WAV format...")
                    audio = AudioSegment.from_file(audio_path)
                    audio.export(audio_path, format="wav")

            # Load the audio
            y, sr = librosa.load(audio_path, sr=22050)
            mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
            mfccs_mean = np.mean(mfccs, axis=1)

            features.append(mfccs_mean)
            valid_ids.append(candidate_id)

        except Exception as e:
            print(f"Skipping {candidate_id}: {e}")

    return np.array(features), valid_ids


filtered_sounds_path = "cough_only"
X_train_audio, id_train_valid = extract_audio_features(y_train_ex.index, filtered_sounds_path)
X_test_audio, id_test_valid = extract_audio_features(y_test_ex, filtered_sounds_path)

print("Final Train Features Shape:", X_train_audio.shape)
print("Final Test Features Shape:", X_test_audio.shape)
```

```python
source_folder = "cough_only"         
output_folder = "spectograms"        
os.makedirs(output_folder, exist_ok=True)

def convert_to_melspectrogram(candidate_id):
    """Convert a WAV file to a Mel spectrogram image and save it."""
    file_name = f"{candidate_id}.wav"
    file_path = os.path.join(source_folder, file_name)

    if not os.path.exists(file_path):
        print(f"File not found: {file_path}")
        return

    y, sr = librosa.load(file_path, sr=22050)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

    output_image = os.path.join(output_folder, f"{candidate_id}.png")
    plt.figure(figsize=(4, 4))
    librosa.display.specshow(mel_spec_db, sr=sr, x_axis="time", y_axis="mel", cmap="viridis")
    plt.axis("off")
    plt.savefig(output_image, bbox_inches="tight", pad_inches=0)
    plt.close()

    print(f"Converted {file_name}")

candidate_ids = y_train['candidateID'].tolist() 
for candidate_id in candidate_ids:                        
    convert_to_melspectrogram(candidate_id) 
```

```python
# Extract MFCC Sequences for audio files
def extract_mfcc_sequence(file_path, n_mfcc=13, max_pad_len=40):
    y, sr = librosa.load(file_path, sr=22050)  
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)  
    mfcc = mfcc.T
    if mfcc.shape[0] < max_pad_len:
        pad_width = max_pad_len - mfcc.shape[0]
        mfcc = np.pad(mfcc, ((0, pad_width), (0, 0)), mode='constant')
    else:
        mfcc = mfcc[:max_pad_len, :]
    return mfcc  


mfcc_sequences = []
ids = []
candidate_ids = y_train['candidateID'].tolist() 
for cid in candidate_ids:
    path = f"cough_only/{cid}.wav"  
    if os.path.exists(path):
        mfcc_seq = extract_mfcc_sequence(path, n_mfcc=13, max_pad_len=40)
        mfcc_sequences.append(mfcc_seq)
        ids.append(cid)
mfcc_sequences = np.array(mfcc_sequences)  

```

#### CNN + ANN using MFCC Sequences

```python

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Flatten, Dense,
    Dropout, Concatenate
)
from tensorflow.keras.utils import to_categorical

num_classes = 3
y_train_final = to_categorical(y_train_final, num_classes)
X_audio_cnn = mfcc_sequences.reshape((mfcc_sequences.shape[0], 40, 13, 1))  

def cnn_mfcc(audio_shape, clinical_shape, num_classes):
    audio_input = Input(shape=audio_shape, name="audio_input")
    x = Conv2D(32, (3, 3), activation="relu", padding="same")(audio_input)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.3)(x)
    x = Conv2D(64, (3, 3), activation="relu", padding="same")(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.3)(x)
    x = Flatten()(x)

    
    clinical_input = Input(shape=(clinical_shape,), name="clinical_input")
    y = Dense(64, activation="relu")(clinical_input)
    y = Dropout(0.3)(y)
    y = Dense(32, activation="relu")(y)

    
    merged = Concatenate()([x, y])
    z = Dense(64, activation="relu")(merged)
    z = Dropout(0.3)(z)
    output = Dense(num_classes, activation="softmax")(z)

    return Model(inputs=[audio_input, clinical_input], outputs=output)
```

```python
model = cnn_mfcc(
    audio_shape=(X_audio_cnn.shape[1], X_audio_cnn.shape[2], 1),  
    clinical_shape=train_clinical_data.shape[1],  
    num_classes=3
)
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

history = model.fit(
    [X_audio_cnn, train_clinical_data],
    y_train_final,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

#### Resnet-50 using MFCC Sequences

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import (
    BatchNormalization, Activation, Add, GlobalAveragePooling2D
)

def identity_block(x, filters):
    f1, f2, f3 = filters
    x_short = x
    x = Conv2D(f1, (1, 1))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f2, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f3, (1, 1))(x)
    x = BatchNormalization()(x)

    x = Add()([x, x_short])
    x = Activation('relu')(x)
    return x

def conv_block(x, filters, strides=(2, 2)):
    f1, f2, f3 = filters
    x_short = Conv2D(f3, (1, 1), strides=strides)(x)
    x_short = BatchNormalization()(x_short)

    x = Conv2D(f1, (1, 1), strides=strides)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f2, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f3, (1, 1))(x)
    x = BatchNormalization()(x)

    x = Add()([x, x_short])
    x = Activation('relu')(x)
    return x

def ResNet50_audio_branch(input_tensor):
    x = Conv2D(64, (7, 7), strides=(2, 2), padding='same')(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)
    
    x = conv_block(x, [64, 64, 256], strides=(1, 1))
    x = identity_block(x, [64, 64, 256])
    x = identity_block(x, [64, 64, 256])
    
    x = conv_block(x, [128, 128, 512])
    x = identity_block(x, [128, 128, 512])
    x = identity_block(x, [128, 128, 512])
    x = identity_block(x, [128, 128, 512])

    x = conv_block(x, [256, 256, 1024])
    for _ in range(5):
        x = identity_block(x, [256, 256, 1024])

    x = conv_block(x, [512, 512, 2048])
    x = identity_block(x, [512, 512, 2048])
    x = identity_block(x, [512, 512, 2048])

    x = GlobalAveragePooling2D()(x)
    return x


def ClinicalBranch(clinical_input):
    x = Dense(128, activation='relu')(clinical_input)
    x = Dense(64, activation='relu')(x)
    x = Dense(32, activation='relu')(x)
    return x


def Resnet50_mfcc(audio_input_shape, clinical_input_shape, num_classes):
    audio_input = Input(shape=audio_input_shape, name="audio_input")
    clinical_input = Input(shape=clinical_input_shape, name="clinical_input")

    audio_features = ResNet50_audio_branch(audio_input)
    clinical_features = ClinicalBranch(clinical_input)

    merged = Concatenate()([audio_features, clinical_features])
    x = Dense(128, activation='relu')(merged)
    x = Dense(64, activation='relu')(x)
    output = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=[clinical_input, audio_input], outputs=output)
    return model
```

```python
audio_shape = (40, 13, 1)         
clinical_shape = (9,)           
num_classes = 3

X_train_audio = np.expand_dims(mfcc_sequences, axis=-1)  
X_train_clinical = np.array(train_clinical_data).astype(np.float32)  
assert X_train_audio.shape[0] == X_train_clinical.shape[0] == y_train_final.shape[0]

model = Resnet50_mfcc(audio_shape, clinical_shape, num_classes)
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

model.fit(
    [X_train_clinical, X_train_audio],
    y_train_final,
    validation_split=0.2,
    epochs=50,
    batch_size=32
)
```

#### Biresnet using MFCC Sequences

```python
def identity_block_2d(x, filters):
    f1, f2, f3 = filters
    x_short = x

    x = Conv2D(f1, (1, 1))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f2, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f3, (1, 1))(x)
    x = BatchNormalization()(x)

    x = Add()([x, x_short])
    x = Activation('relu')(x)
    return x

def conv_block_2d(x, filters, strides=(2, 2)):
    f1, f2, f3 = filters
    x_short = Conv2D(f3, (1, 1), strides=strides)(x)
    x_short = BatchNormalization()(x_short)

    x = Conv2D(f1, (1, 1), strides=strides)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f2, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f3, (1, 1))(x)
    x = BatchNormalization()(x)

    x = Add()([x, x_short])
    x = Activation('relu')(x)
    return x

def ResNet50_audio_branch(input_tensor):
    x = Conv2D(64, (7, 7), strides=(2, 2), padding='same')(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    x = conv_block_2d(x, [64, 64, 256], strides=(1, 1))
    x = identity_block_2d(x, [64, 64, 256])
    x = identity_block_2d(x, [64, 64, 256])

    x = conv_block_2d(x, [128, 128, 512])
    for _ in range(3):
        x = identity_block_2d(x, [128, 128, 512])

    x = conv_block_2d(x, [256, 256, 1024])
    for _ in range(5):
        x = identity_block_2d(x, [256, 256, 1024])

    x = conv_block_2d(x, [512, 512, 2048])
    x = identity_block_2d(x, [512, 512, 2048])
    x = identity_block_2d(x, [512, 512, 2048])

    x = GlobalAveragePooling2D()(x)
    return x


def ResNet_clinical_branch(input_tensor):
    x = Dense(64)(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    skip = x
    x = Dense(64)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Dense(64)(x)
    x = BatchNormalization()(x)

    x = Add()([x, skip])
    x = Activation('relu')(x)

    x = Dense(32, activation='relu')(x)
    return x


def BiResNet_mfcc(audio_input_shape, clinical_input_shape, num_classes):
    audio_input = Input(shape=audio_input_shape, name="audio_input")
    clinical_input = Input(shape=clinical_input_shape, name="clinical_input")

    audio_features = ResNet50_audio_branch(audio_input)
    clinical_features = ResNet_clinical_branch(clinical_input)

    merged = Concatenate()([audio_features, clinical_features])
    x = Dense(128, activation='relu')(merged)
    x = Dense(64, activation='relu')(x)
    output = Dense(num_classes, activation='softmax')(x)

    return Model(inputs=[clinical_input, audio_input], outputs=output)
```

```python
audio_shape = (40, 13, 1)  
clinical_shape = (9,)      
num_classes = 3

X_train_audio = np.expand_dims(mfcc_sequences, axis=-1)  
X_train_clinical = np.array(train_clinical_data).astype(np.float32)  
y_train = np.array(y_train_final).astype(np.int32)

model = BiResNet_mfcc(audio_shape, clinical_shape, num_classes)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(
    [X_train_clinical, X_train_audio],
    y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32
)
```

#### CNN + ANN using Spectrograms

```python
def cnn_spect(spectrogram_shape, clinical_shape, num_classes):
    spectrogram_input = Input(shape=spectrogram_shape, name="spectrogram_input")
    x = Conv2D(32, (3, 3), activation='relu')(spectrogram_input)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.3)(x)
    
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.3)(x)
    
    x = Flatten()(x)  

    
    clinical_input = Input(shape=(clinical_shape,), name="clinical_input")
    y = Dense(64, activation='relu')(clinical_input)
    y = Dropout(0.3)(y)
    y = Dense(32, activation='relu')(y)

    
    merged = Concatenate()([x, y])
    z = Dense(64, activation='relu')(merged)
    z = Dropout(0.3)(z)
    output = Dense(num_classes, activation='softmax')(z)

    return Model(inputs=[spectrogram_input, clinical_input], outputs=output)
```

#### Resnet-50 using Spectrograms

```python
def ResNet50_spect(audio_input_shape, clinical_input_shape, num_classes):
    audio_input = Input(shape=audio_input_shape, name="audio_input")
    if audio_input_shape[-1] == 1:
        x = tf.keras.layers.Concatenate()([audio_input, audio_input, audio_input])
    else:
        x = audio_input


    resnet_base = ResNet50(weights=None, include_top=False, input_tensor=x)
    x = resnet_base.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.3)(x)

   
    clinical_input = Input(shape=clinical_input_shape, name="clinical_input")
    y = Dense(64, activation='relu')(clinical_input)
    y = Dropout(0.3)(y)
    y = Dense(32, activation='relu')(y)

    
    combined = Concatenate()([x, y])
    z = Dense(128, activation='relu')(combined)
    z = Dropout(0.3)(z)
    output = Dense(num_classes, activation='softmax')(z)

    return Model(inputs=[audio_input, clinical_input], outputs=output)
```

#### Biresnet + ANN using spectrograms

```python
def res_block(x, filters, strides=(1,1)):
    shortcut = x
    f1, f2, f3 = filters

    x = Conv2D(f1, (1,1), strides=strides, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f2, (3,3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f3, (1,1), padding='same')(x)
    x = BatchNormalization()(x)

    if strides != (1,1) or shortcut.shape[-1] != f3:
        shortcut = Conv2D(f3, (1,1), strides=strides, padding='same')(shortcut)
        shortcut = BatchNormalization()(shortcut)

    x = Add()([x, shortcut])
    x = Activation('relu')(x)
    return x

def BiResNetAudioBranch(input_tensor):
    x = Conv2D(64, (7,7), strides=(2,2), padding='same')(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((3,3), strides=(2,2), padding='same')(x)

    x = res_block(x, [64,64,256], strides=(1,1))
    x = res_block(x, [64,64,256])
    x = res_block(x, [64,64,256])

    x = res_block(x, [128,128,512], strides=(2,2))
    x = res_block(x, [128,128,512])
    x = res_block(x, [128,128,512])

    x = res_block(x, [256,256,1024], strides=(2,2))
    for _ in range(4):
        x = res_block(x, [256,256,1024])

    x = res_block(x, [512,512,2048], strides=(2,2))
    x = res_block(x, [512,512,2048])
    x = res_block(x, [512,512,2048])

    x = GlobalAveragePooling2D()(x)
    return x

def ClinicalANNBranch(clinical_input):
    x = Dense(128, activation='relu')(clinical_input)
    x = Dropout(0.3)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(32, activation='relu')(x)
    return x

def BiResNet_spect(audio_input_shape, clinical_input_shape, num_classes):
    audio_input = Input(shape=audio_input_shape, name="audio_input")
    clinical_input = Input(shape=clinical_input_shape, name="clinical_input")

    audio_features = BiResNetAudioBranch(audio_input)
    clinical_features = ClinicalANNBranch(clinical_input)

    merged = Concatenate()([audio_features, clinical_features])
    x = Dense(256, activation='relu')(merged)
    x = Dropout(0.4)(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.4)(x)
    output = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=[clinical_input, audio_input], outputs=output)
    return model
```

```python
audio_shape = (128, 128, 1)                         
clinical_shape = (train_clinical_data.shape[1],)       
num_classes = 3  

model = BiResNet_spect(audio_shape, clinical_shape, num_classes)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    [train_clinical_data, X_train_spectrogram],
    y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32
)
```

#### AST + ANN

```python
import torchaudio
from transformers import ASTFeatureExtractor, ASTModel
from tqdm import tqdm

# Load AST model and extractor
feature_extractor = ASTFeatureExtractor.from_pretrained("MIT/ast-finetuned-audioset-10-10-0.4593")
ast_model = ASTModel.from_pretrained("MIT/ast-finetuned-audioset-10-10-0.4593")
ast_model.eval()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
ast_model.to(device)
```

```python
train_wav_files = [
    os.path.join("cough_only", f"{str(cid)}.wav")
    for cid in X_train_filtered["candidateID"]
]
```

```python
# Converting raw audio files into waveform tensors
def load_waveform(path, target_sr=16000):
    try:
        waveform, sr = torchaudio.load(path)
        if waveform.numel() == 0:
            print(f"Empty audio in file: {path}")
            return None
        waveform = waveform.mean(dim=0)  # mono
        if sr != target_sr:
            waveform = torchaudio.functional.resample(waveform, sr, target_sr)
        return waveform
    except Exception as e:
        print(f"Failed to load {path}: {e}")
        return None


waveform_list = []
bad_paths = []
for path in tqdm(train_wav_files):
    w = load_waveform(path)
    if w is not None:
        waveform_list.append(w)
    else:
        bad_paths.append(path)

print(f"\nLoaded: {len(waveform_list)}")
print(f"\n Failed: {len(bad_paths)}")

target_len = 16000 * 5
processed_list = []
for w in waveform_list:
    if w.shape[0] > target_len:
        w = w[:target_len]
    elif w.shape[0] < target_len:
        w = torch.nn.functional.pad(w, (0, target_len - w.shape[0]))
    processed_list.append(w)

waveform_tensor = torch.stack(processed_list)
print("Final tensor shape:", waveform_tensor.shape)
```

```python
waveform_tensor
```

```python
def get_ast_embeddings(waveforms, batch_size=16):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    ast_model.to(device)
    all_embeddings = []
    for i in range(0, len(waveforms), batch_size):
        batch = waveforms[i:i + batch_size]
        batch_np = [w.cpu().numpy() for w in batch]
        inputs = feature_extractor(
            batch_np,
            sampling_rate=16000,
            return_tensors="pt"
        ).input_values.to(device)

        with torch.no_grad():
            outputs = ast_model(inputs)
            pooled = outputs.pooler_output  
            all_embeddings.append(pooled.cpu().numpy())
    return np.vstack(all_embeddings)

X_train_ast = get_ast_embeddings(waveform_tensor)
```

```python
ast_input = Input(shape=(768,), name="ast_input")
x = Dense(256, activation="relu")(ast_input)
x = Dropout(0.3)(x)


clinical_input = Input(shape=(9,), name="clinical_input")
y = Dense(64, activation="relu")(clinical_input)
y = BatchNormalization()(y)
y = Dense(64, activation="relu")(y)


fusion = Concatenate()([x, y])
fusion = Dense(128, activation='relu')(fusion)
fusion = Dropout(0.4)(fusion)
fusion = Dense(64, activation='relu')(fusion)
output = Dense(3, activation='softmax')(fusion)


model = Model(inputs=[ast_input, clinical_input], outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_accuracy',   
    patience=1,                
    mode='max',                
    restore_best_weights=True  
)

model.fit(
    [X_train_ast, train_clinical_data],
    y_train_final,
    validation_split=0.3,
    epochs=12,
    batch_size=32,
    callbacks=[early_stop]
)
```
